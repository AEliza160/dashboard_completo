```python
import pandas as pd
import matplotlib.pyplot as plt
from google.colab import files

# Subir y cargar archivo (móviles)
try:
    df = pd.read_csv("/content/train.csv", encoding="utf-8")
except:
    print("Sube el archivo:")
    uploaded = files.upload()
    df = pd.read_csv("/content/train.csv", encoding="utf-8")

# Configurar figura grande para todos los gráficos
plt.figure(figsize=(24,16))

# 1. Rango de precios más comunes
plt.subplot(2,3,1)
df['price_range'].value_counts().plot(kind='bar', color='skyblue')
plt.title("Rangos de precio", fontsize=18)
plt.xlabel("Rango", fontsize=14)
plt.ylabel("Cantidad de móviles", fontsize=14)
plt.xticks(rotation=0, fontsize=12)
plt.yticks(fontsize=12)

# 2. Histograma de batería
plt.subplot(2,3,2)
df['battery_power'].plot(kind='hist', bins=20, color='green')
plt.title("Distribución de batería", fontsize=18)
plt.xlabel("mAh", fontsize=14)
plt.ylabel("Cantidad", fontsize=14)
plt.xticks(fontsize=12)
plt.yticks(fontsize=12)

# 3. RAM de los móviles
plt.subplot(2,3,3)
df['ram'].plot(kind='hist', bins=20, color='orange')
plt.title("Distribución de RAM", fontsize=18)
plt.xlabel("RAM (MB)", fontsize=14)
plt.ylabel("Cantidad de móviles", fontsize=14)
plt.xticks(fontsize=12)
plt.yticks(fontsize=12)

# 4. Móviles con más memoria interna
plt.subplot(2,3,4)
df['int_memory'].value_counts().head(10).plot(kind='bar', color='purple')
plt.title("Top 10 memoria interna", fontsize=18)
plt.xlabel("Memoria interna (GB)", fontsize=14)
plt.ylabel("Cantidad", fontsize=14)
plt.xticks(rotation=45, fontsize=12)
plt.yticks(fontsize=12)

# 5. Batería promedio por rango de precio
plt.subplot(2,3,5)
df.groupby('price_range')['battery_power'].mean().sort_values().plot(kind='barh', color='pink')
plt.title("Batería promedio por rango de precio", fontsize=18)
plt.xlabel("mAh", fontsize=14)
plt.ylabel("Rango de precio", fontsize=14)
plt.xticks(fontsize=12)
plt.yticks(fontsize=12)

# 6. RAM promedio vs rango de precio (línea)
plt.subplot(2,3,6)
ram_avg = df.groupby('price_range')['ram'].mean()
plt.plot(ram_avg.index, ram_avg.values, marker='o', color='red')
plt.title("RAM promedio por rango de precio", fontsize=18)
plt.xlabel("Rango de precio", fontsize=14)
plt.ylabel("RAM promedio (MB)", fontsize=14)
plt.grid(True)
plt.xticks(rotation=0, fontsize=12)
plt.yticks(fontsize=12)

# Ajustar espacio
plt.subplots_adjust(hspace=0.45, wspace=0.35)
plt.tight_layout(pad=2.0)

# Guardar imagen en alta calidad
plt.savefig("dashboard_mobiles.png", dpi=300, bbox_inches="tight")

# Mostrar gráfico
plt.show()

# Descargar la imagen (Colab)
files.download("dashboard_mobiles.png")
```
